```python
import pandas as pd
import numpy as np
```

```python
table = pd.read_csv("/content/drive/MyDrive/DSC 341 (LAB)/Video_Store.csv", sep=',')
table
```

```python
def min_max_scaling(data, new_min=0, new_max=1):
    return (data - np.min(data)) / (np.max(data) - np.min(data)) * (new_max - new_min) + new_min
```

```python
table[['Income', 'Age', 'Rentals']].apply(min_max_scaling)
```

```python
table
```

```python
table[['norm_Income', 'norm_Age', 'norm_Rentals']] = table[['Income', 'Age', 'Rentals']].apply(min_max_scaling)
```

```python
table
```

```python
crosstab = pd.crosstab(table.Gender,table.Genre)
crosstab
```

```python
import matplotlib.pyplot as plt
```

```python
crosstab = pd.crosstab(table.Gender, table.Genre)
```

```python

fig, ax = plt.subplots(1,2, figsize =(15,5))
crosstab.plot(kind='bar', ax = ax[0])
crosstab.T.plot(kind='bar', ax= ax[1])
ax[0].set_title("Movie GenreCounts by Gender", fontsize = 15)
ax[1].set_title("Gender Counts by Movie Genre", fontsize = 15)
ax[0].set_ylabel('Count', fontsize = 10)
ax[1].set_ylabel('Count', fontsize = 10)
ax[0].set_xlabel('Gender', fontsize = 10)
ax[1].set_xlabel('Gender', fontsize = 10)
plt.show()
```

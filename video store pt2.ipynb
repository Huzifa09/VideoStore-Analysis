```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/data science assignment 3/Video_Store_2.csv")
df.head(5)
```

```python
df_target = df.Incidentals
```

```python
df = pd.get_dummies(df[['Gender','Income','Age','Rentals','Avg Per Visit','Genre']])
df.head(5)
```

```python
tpercent = 0.8
tsize = int(tpercent * len(df))
df_train = df[:tsize]
df_test = df[tsize:]
print(df_train.shape)
print(df_test.shape)
```

```python
from sklearn import preprocessing 
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
min_max_scaler.fit(df_train)
```

```python
df_train_norm = min_max_scaler.fit_transform(df_train)
df_test_norm = min_max_scaler.fit_transform(df_test)
```
